```python
import mysql.connector
import pandas as pd

# MySQL Connection Details
db_config = {
    "host": "127.0.0.1",
    "user": "root",
    "password": "1234", 
    "database": "SocialMediaDB"
}

try:
    conn = mysql.connector.connect(**db_config)
    cursor = conn.cursor()
    print("MySQL Connection Successful!")

    # Load Dataset
    df = pd.read_csv('instagram_usage_lifestyle.csv')
    df = df.drop_duplicates().fillna(0)

    # Randomly select 1200 records
    sample_df = df.sample(n=1200)

    for _, row in sample_df.iterrows():
        # Insertion logic according to FK constraints
        # 1. UserAccount
        cursor.execute("INSERT IGNORE INTO UserAccount VALUES (%s, %s, %s, %s)", 
                       (row['user_id'], row['account_creation_year'], row['subscription_status'], row['uses_premium_features']))
        
        # 2. UserDemographics
        cursor.execute("INSERT IGNORE INTO UserDemographics VALUES (%s, %s, %s, %s, %s, %s, %s, %s)",
                       (row['user_id'], row['age'], row['gender'], row['country'], row['urban_rural'], row['income_level'], row['employment_status'], row['relationship_status']))
        
        # 3. SecuritySettings
        cursor.execute("INSERT IGNORE INTO SecuritySettings VALUES (%s, %s, %s, %s, %s)",
                       (row['user_id'], row['privacy_setting_level'], row['two_factor_auth_enabled'], row['biometric_login_used'], row['linked_accounts_count']))
        
        # 4. SocialEngagement
        cursor.execute("INSERT IGNORE INTO SocialEngagement VALUES (%s, %s, %s, %s, %s, %s, %s)",
                       (row['user_id'], row['followers_count'], row['following_count'], row['preferred_content_theme'], row['user_engagement_score'], row['ads_viewed_per_day'], row['ads_clicked_per_day']))
        
        # 5. AppUsage
        cursor.execute("INSERT IGNORE INTO AppUsage VALUES (%s, %s, %s, %s, %s, %s, %s, %s)",
                       (row['user_id'], row['daily_active_minutes_instagram'], row['sessions_per_day'], row['posts_created_per_week'], row['stories_viewed_per_day'], row['dms_sent_per_week'], row['dms_received_per_week'], row['notification_response_rate']))
        
        # 6. HealthLifestyle
        cursor.execute("INSERT IGNORE INTO HealthLifestyle VALUES (%s, %s, %s, %s, %s, %s, %s)",
                       (row['user_id'], row['exercise_hours_per_week'], row['sleep_hours_per_night'], row['smoking'], row['alcohol_frequency'], row['perceived_stress_score'], row['self_reported_happiness']))

    conn.commit()
    print(f"Data Insertion Complete: {len(sample_df)} records added.")

except mysql.connector.Error as err:
    print(f"Error: {err}")
finally:
    if 'conn' in locals() and conn.is_connected():
        cursor.close()
        conn.close()
```
